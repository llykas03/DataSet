```python
#bolje razumevanje kolone gde je tip podatka object
def open_object_dtype(s):
    dtypes = set()
    dtypes.update(s.apply(type))
    return dtypes

manga_df.apply(open_object_dtype, axis=0)
```

```python
def describe(manga_df):

    variables = []
    dtypes = []
    count = []
    unique = []
    missing = []
    min_ = []
    max_ = []

    for item in manga_df.columns:
        variables.append(item)
        dtypes.append(manga_df[item].dtype)
        count.append(len(manga_df[item]))
        unique.append(len(manga_df[item].unique()))
        missing.append(manga_df[item].isna().sum())
        
        if manga_df[item].dtypes == 'float64' or manga_df[item].dtypes == 'int64':
            min_.append(manga_df[item].min())
            max_.append(manga_df[item].max())
        else:
            min_.append('Str')
            max_.append('Str')



    output = pd.DataFrame({
        'variable': variables, 
        'dtype': dtypes,
        'count': count,
        'unique': unique,
        'missing value': missing,
        'Min': min_,
        'Max': max_
    })    
        
    return output
```

```python
desc_df = describe(manga_df)
desc_df = desc_df.style.set_caption('Overview of the dataset').set_table_styles(Styles)
desc_df.set_properties(**{'background-color': '#F6E7E8',
                           'color': 'black',
                           'border-color': 'black'})
```
